## Invers to the power of the distance
This notebook is to prototype an estimator, starting with something simple as IDp. The objective here is to define a search strategy.

```python
import numpy as np
from matplotlib import pyplot as plt
from search2 import Neighborhood
from estimator import Estimator
%matplotlib inline
```

```python
# generate some fake data 
np.random.seed(0)
x = np.random.rand(100, 3) # a 3D dataset
x[:,2]=0 # make it 2D
x0 = np.random.rand(10000) # targets
y0 = np.random.rand(10000) # targets
z0 = np.zeros(10000) # targets
plt.plot(x0,y0,'.')
plt.plot(x[:,0],x[:,1],'.')
```

```python
n = Neighborhood(x[:,0],x[:,1],x[:,2], a = [[0.3,0.3,0.3],[0.3,0.3,0.3]], rot=[[45,0,0],[-45,0,0]])
```

```python
e =Estimator(x0, y0, z0, search = n)
```

```python
e.mean()
```
