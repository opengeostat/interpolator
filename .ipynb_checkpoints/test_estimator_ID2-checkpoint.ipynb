# Invers to the power of the distance

```python
import numpy as np
from matplotlib import pyplot as plt
from neighborhood import Neighborhood
from estimator import Estimator
%matplotlib inline
```

```python
# generate some fake data 
np.random.seed(0)
x = np.random.rand(100, 3) # a 3D dataset
x[:,2]=0 # make it 2D
v =  np.random.rand(100, 2) # two variables 

# generate target (a grid)
nx, ny, nz = (20, 20, 1)
xt = np.linspace(0, 1, nx)
yt = np.linspace(0, 1, ny)
zt = np.linspace(0, 0, nz)
x0, y0, z0 = np.meshgrid(xt, yt, zt)


plt.plot(x0.ravel(),y0.ravel(),'.')
plt.scatter(x[:,0],x[:,1],c = v[:,0], edgecolors = 'k')
```

```python
n = Neighborhood(x[:,0],x[:,1],x[:,2], a = [[0.3,0.1,0.1],[0.3,0.3,0.3]], rot=[[45,0,0],[-45,0,0]], 
                 v = v, vname = ['v1','v2'], vmeta = ['a random variable', 'a random variable'])
```

```python
e =Estimator(x0.ravel(), y0.ravel(), z0.ravel(), search = n)
```

```python
e.id_power(debug = True, power = 3)
```

```python
e.estimates['e1']['estimate'][0:2]
```

```python
#plt.scatter(e.x0,e.y0, s = 0.6,c=e.estimates['e1']['estimate'][:,0])
variable_id = 0
plt.contourf(xt,yt,e.estimates['e1']['estimate'][:,variable_id ].reshape([nx,ny]))
plt.plot(n.x,n.y,'+k')
plt.axis('equal');
plt.axis('scaled');
plt.colorbar()
plt.title(e.estimates['e1']['meta'] +" "+ e.estimates['e1']['vname'][variable_id]);
```

```python
#plt.scatter(e.x0,e.y0, s = 0.6,c=e.estimates['e1']['estimate'][:,0])
variable_id = 1
plt.contourf(xt,yt,e.estimates['e1']['estimate'][:,variable_id ].reshape([nx,ny]))
plt.plot(n.x,n.y,'+k')
plt.axis('equal');
plt.axis('scaled');
plt.colorbar()
plt.title(e.estimates['e1']['meta'] +" "+ e.estimates['e1']['vname'][variable_id]);
```
