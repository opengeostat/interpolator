## Invers to the power of the distance
This notebook is to prototype an estimator, starting with something simple as IDp. The objective here is to define a search strategy.

```python
import numpy as np
from matplotlib import pyplot as plt
#import concurrent.futures
from multiprocessing import Pool
from search import Neighborhood
%matplotlib inline
```

```python
# generate some fake data 
np.random.seed(0)
x = np.random.rand(10000, 3)+10 # a 3D dataset
x[:,2]=0 # make it 2D
x0 = np.array([0.5,0.5,0.5])+10 # origing of coordinate
plt.plot(x[:,0],x[:,1],'.')
plt.plot(x0[0],x0[1],'o')
```

```python
n = Neighborhood(x[:,0],x[:,1],x[:,2], a = [[0.3,0.1,0.1],[0.3,0.1,0.1]], rot=[[45,0,0],[-45,0,0]])
```

```python
n.update(np.array([10.5,10.5,0]))
```

```python
plt.scatter(n.x,n.y,c=np.sqrt(n.sqdisti))
plt.plot(n.x[n.test],n.y[n.test],'.r',alpha = 0.1)
plt.plot(n.x0,n.y0,'o')
plt.axis('equal');
plt.axis('scaled');
plt.colorbar()
plt.title('isotropic distance^2 ')
```

```python
plt.plot(n.xi,n.yi,'.')
plt.plot(n.xi[n.test],n.yi[n.test],'.')
plt.plot(n.x0i,n.y0i,'o')
plt.axis('equal');
#plt.axis('scaled');
```

```python
plt.scatter(n.x[n.test],n.y[n.test],c=n.get_distances())
plt.axis('equal');
plt.axis('scaled');
plt.title('anisotropic distance')
plt.colorbar()
```

```python
plt.scatter(n.x[n.test],n.y[n.test],c=n.get_distances(0))
plt.axis('equal');
plt.axis('scaled');
plt.title('isotropic distance 0')
plt.colorbar()
```

```python
# the angle of the second rotation is inverted: rot=[[45,0,0],[-45,0,0]]
plt.scatter(n.x[n.test],n.y[n.test],c=n.get_distances(1))
plt.axis('equal');
plt.axis('scaled');
plt.title('isotropic distance 1')
plt.colorbar()
```
